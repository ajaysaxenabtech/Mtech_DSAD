```python
# Cycle Startup assignment
# Group 050
#################
Data_Given = []
Queries = []
with open("inputPS08.txt") as Input_file:
    for a_line in Input_file:
        if a_line.lower().startswith("park"):
            Data_Given.append(a_line.lower().strip())
        else:
            Queries.append(a_line.lower().strip())
print(Data_Given)
print(Queries)
print("\n\n")

def heapify(input_array,size_of_input_array,idx):
    largest = idx
    left_child_position = 2 * idx + 1
    right_child_position = 2 * idx + 2
    if left_child_position < size_of_input_array and input_array[idx] <= input_array[left_child_position]:
        largest = left_child_position
    if right_child_position < size_of_input_array and input_array[largest] <= input_array[right_child_position]:
        largest = right_child_position
    if largest != idx:
        input_array[idx],input_array[largest] = input_array[largest],input_array[idx]
        heapify(input_array,size_of_input_array,largest)

def heap_sort(input_array):
    length_of_array = len(input_array)
    print(input_array)
    for a_position in range(length_of_array // 2, -1, -1):
        heapify(input_array,length_of_array,a_position)
    print(input_array)
    for a_position in range(length_of_array-1,0,-1):
        input_array[0],input_array[a_position] = input_array[a_position],input_array[0]
        heapify(input_array,a_position,0)
    print(input_array)
    print(input_array[-1])
        
def list_all_parks(Data_Given):
    parks = []
    for an_entry in Data_Given:
        if an_entry.split(",")[0].capitalize() not in parks:
            parks.append(an_entry.split(",")[0].capitalize())
        else:
            pass
    print("List of cycling parks:")
    print(",".join(parks),"\n\n")

def cycles_per_month(Data_Given):
    months = []
    current_month_number_of_cycles = 0
    print("Total number of cycles required per month:")
    for an_entry in Data_Given:
        if len(months)==0:
            months.append(an_entry.split(",")[-1].strip())
            current_month_number_of_cycles = current_month_number_of_cycles + int(an_entry.split(",")[1])
        elif (an_entry.split(",")[-1].strip() in months) and len(months)!=0:
            current_month_number_of_cycles = current_month_number_of_cycles + int(an_entry.split(",")[1])
        elif (an_entry.split(",")[-1].strip() not in months) and len(months)!=0:
            print(months[-1].capitalize(),current_month_number_of_cycles,sep=" ")
            months = []
            current_month_number_of_cycles = 0
            months.append(an_entry.split(",")[-1].strip())
            current_month_number_of_cycles = current_month_number_of_cycles + int(an_entry.split(",")[1])
    print(months[-1].capitalize(),current_month_number_of_cycles,sep=" ",end="")
    print("\n\n")

def max_cycles_per_month(Data_Given,Month):
    Month = Month.lower().strip()
    cycles_used_that_month = []
    for an_entry in Data_Given:
        if an_entry.split(",")[-1].strip().lower()==Month:
            cycles_used_that_month.append(int(an_entry.split(",")[1].strip()))
    heap_sort(cycles_used_that_month)

def add_new_park(Data_Given,Month):
    return Data_Given.append(Month.lower())

def remove_a_park(Data_Given,park_name):
    try:
        for an_entry in Data_Given:
            if an_entry.split(",")[0].strip().lower()==park_name:
                  Data_Given.remove(an_entry)
            break
    except ValueError:
        pass
    
def print_max_heap_tree_for_each_month(Data_Given):
    print("Max Heap Tree for Each Month")
    Months = []
    current_month_data = []
    for an_entry in Data_Given:
        if an_entry.split(",")[-1].strip().lower() not in Months:
            Months.append(an_entry.split(",")[-1].strip().lower())
    for a_month in Months:
        for an_entry in Data_Given:
            if an_entry.split(",")[-1].strip().lower() == a_month:
                current_month_data.append(int(an_entry.split(",")[1].strip()))
        print(current_month_data)
        length_of_array = len(current_month_data)
        for a_position in range(length_of_array // 2, -1, -1):
               heapify(current_month_data,length_of_array,a_position)
        print("Max-Heap Tree for the month of {Month}: ".format(Month=a_month.capitalize()))
        for i in range(len(current_month_data)):
            print("%d "%current_month_data[i],end="")
        print()
        current_month_data = []

for a_query in Queries:
    if a_query == "listallparks":
        list_all_parks(Data_Given)
    elif a_query == "getcyclespermonth":
        cycles_per_month(Data_Given)
    elif a_query.startswith("maxcycles"):
        max_cycles_per_month(Data_Given,a_query.split(":")[-1].strip())
    elif a_query.startswith("add park"):
        print("ADD_PARK")
        print(len(Data_Given))
        add_new_park(Data_Given,a_query.split(":")[-1].strip())
        print(len(Data_Given))
    elif a_query.startswith("remove park"):
        print("REMOVE PARK")
        print(len(Data_Given))
        remove_a_park(Data_Given,a_query.split(":")[-1].strip())
        print(len(Data_Given))
    elif a_query.startswith("maxheap"):
        print_max_heap_tree_for_each_month(Data_Given)
        
        
        

        
```
